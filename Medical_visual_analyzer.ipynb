```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('medical_examination.csv')
df.head(10)
```

```python
df['BMI'] = df['weight'] / ((df['height']/100)**2)
df.head(10)
```

```python
def over_weight(x):
    if x > 25:
        return 1
    else:
        return 0

df['Overweight'] = df['BMI'].apply(over_weight)
```

```python
df.head()
```

```python
def norm_lize(x):
  if x > 1:
    return 1
  else:
    return 0

df['cholesterol'] = df['cholesterol'].apply(norm_lize)
df['gluc'] = df['gluc'].apply(norm_lize)
```
